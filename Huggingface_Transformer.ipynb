### Pipeline

```python
from transformers import pipeline

classifier = pipeline('sentiment-analysis',device='cpu')
classifier('The Match was thrilling')
```

```python
classifier('The food was very bad')
```

```python
classifier('The food is neither good nor bad')
```

```python
# https://huggingface.co/cardiffnlp/twitter-roberta-base-sentiment

classifier = pipeline("sentiment-analysis",model="cardiffnlp/twitter-roberta-base-sentiment",device='cpu')

classifier("Messi is the greatest")
```

### Question Answering

```python
question_answerer = pipeline("question-answering",device='cpu')

question_answerer(
    question="Who won the 2023 ballon do or?",
    context="Messi is an Argentine player who won the 2023 ballon do or",
)
```

### Named Entity Recognition

```python
# ner = pipeline("ner",device=0)
ner = pipeline("ner",aggregation_strategy="simple",device='cpu')
ner("Sunil Chetri is an Indian Football Player")
```

### Text Summarization

```python
summarizer = pipeline("summarization",device='cpu')
text = """
    America has changed dramatically during recent years. Not only has the number of
    graduates in traditional engineering disciplines such as mechanical, civil,
    electrical, chemical, and aeronautical engineering declined, but in most of
    the premier American universities engineering curricula now concentrate on
    and encourage largely the study of engineering science. As a result, there
    are declining offerings in engineering subjects dealing with infrastructure,
    the environment, and related issues, and greater concentration on high
    technology subjects, largely supporting increasingly complex scientific
    developments. While the latter is important, it should not be at the expense
    of more traditional engineering.
"""
summarizer(text, max_length=30, min_length=10)
```
